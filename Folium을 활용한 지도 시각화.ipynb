```python
import folium
from folium.plugins import MarkerCluster
from folium.plugins import MiniMap
```

```python
# 지도 생성하기 
m = folium.Map(location = [35.1379222, 129.05562775], zoom_start = 12)

# 현재 지도 m에 MarkerCluster 적용하기 
marker_cluster = MarkerCluster().add_to(m) 

# 미니맵 추가하기 
minimap = MiniMap() 
minimap.add_to(m)

# 부산 명소 마커 추가하기 
for i in range(len(places)):
    folium.Marker([lat[i], lng[i]], tooltip = places[i]).add_to(m) 
m
```
